```python
import mshr
import matplotlib.pyplot as plt
```

```python
R2 = 10.0
R1 = 2.5
outer_circle = mshr.Circle(Point(0, 0), R2)
inner_Circle1 = mshr.Circle(Point(0, 5), R1)
inner_Circle2 = mshr.Circle(Point(0, -5), R1)
inner_Circle3 = mshr.Circle(Point(5, 0), R1)
inner_Circle4 = mshr.Circle(Point(-5, 0), R1)
domain = outer_circle - inner_Circle1 - inner_Circle2 - inner_Circle3 - inner_Circle4
N = int(4 * (R1+R2))
mesh = mshr.generate_mesh(domain, N)
fig = plt.figure(figsize=(8,8))
plot(mesh)
plt.show()
```

```python
# Exploring Different Kinds of Geometries
# ELIPSE
elipse = mshr.Ellipse(Point(0, 0), 10, 5)
circle1 = mshr.Circle(Point(0, 2), 1.5)
circle2 = mshr.Circle(Point(0, -2), 1.5)
circle3 = mshr.Circle(Point(5, 0), 2)
circle4 = mshr.Circle(Point(-5, 0), 2)
domain_elipse = elipse - circle1 - circle2 - circle3 - circle4
filter = 50
mesh_elipse = mshr.generate_mesh(domain_elipse, filter)
fig = plt.figure(figsize=(8,8))
plot(mesh_elipse)
plt.show()
```
