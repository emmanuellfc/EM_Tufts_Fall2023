```python
from __future__ import print_function
import random
from fenics import *
from dolfin import *
import mshr
import matplotlib.pyplot as plt
```

```python
R2 = 10.0

R1 = 4.0
outer_circle = mshr.Circle(Point(0, 0), R2)

inner_Circle = mshr.Circle(Point(0, 0), R1)

domain = outer_circle - inner_Circle
# Create mesh
N = int(3 * (R1+R2))
mesh = mshr.generate_mesh(domain, N)
### Define boundary condition ###
u_D = Constant(0.0)
tol = 1E-14
fig = plt.figure(figsize=(8,8))
plot(mesh)
plt.show()
```
